```python
import pandas as pd
import numpy as np
print("DataFrame.applyfunction:\n")
info=pd.DataFrame([[2,7]]*3,columns=['P','Q'])
print("\nOrigibnal DataFrame:\n",info)
print("\nSquare rppt of DataFrame:\n",info.apply(np.sqrt))
print("\nSum of each column:\n",info.apply(np.sum,axis=0))
print("\nSum of each row:\n",info.apply(np.sum,axis=1))
```
